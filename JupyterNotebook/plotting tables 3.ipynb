## WCO and WOH --> N(OH) vs W(CO) --> N(H2_OH) vs N(H2_CO)

### Equations (Found from Bolatto 2013, Busch 2024, and Busch 2019):

$$ X_{CO} = \frac{N(H_2)}{(W(CO))} $$

$$N(OH) = C\frac{T_{ex}}{T_{ex}-T_c} \int T_b(\nu)d\nu $$ 

where $ \int T_b(\nu)d\nu $ is WOH.

$$ T_{ex} = 5K, T_c = 3K$$

Median Literature Value for N(OH)/N(H2) Ratio: $$\frac{N_{OH}}{N_{H2}} = 1\times 10^{-7}$$

N(H2) from CO: $$N(H_{2, CO}) = X_{CO}W(CO),$$ note this is the J = 0 to 1 transition. 


We use the XCO Factor for the Milky Way from Busch 2024 (via Bolatto 2013): $$X_{CO} = 2\times 10^{20} cm^{-2}$$



```python
#Define Constants:
XCO = 2*10**20
C1667 = 2.3*10**14
#C1665 = same???

#calculate N(OH)
def NOHfinder(Tex, Tc, WOH):
    NOH = C1667*Tex/(Tex-Tc) * WOH
    return NOH

#Calculate NH2_OH
def NH2viaOH(NOH):
    NH2OH = NOH*(10**7)
    return NH2OH

#Calculate NH2_CO
def NH2viaCO(WCO):
    NH2CO = XCO*WCO
    return NH2CO

```

```python
NOH_arr = NOHfinder(5, 3, WOH67)

NH2_OH_arr = NH2viaOH(NOH_arr)

NH2_CO_arr = NH2viaCO(WCO) #Why inverse magnitude?
print(NH2_OH_arr)
print(NH2_CO_arr)
```

```python
plt.scatter(NOH_arr, NH2_CO_arr, color = 'crimson', label = 'NH2_CO', s = 20)
plt.scatter(NOH_arr, NH2_OH_arr, color = 'navy', label = 'NH2_OH', s = 20)
plt.xlabel('NOH [cm^-2]')
plt.ylabel('NH2 [cm^-2]')
plt.legend()
plt.grid(True)
```

```python
# First fit: NH2_OH vs NOH
coeffs1 = np.polyfit(NOH_arr, NH2_OH_arr, 1)  # linear fit
fit_line1 = np.poly1d(coeffs1)
print(f"NH2_OH vs NOH: slope = {coeffs1[0]:.3e}, intercept = {coeffs1[1]:.3e}")

# Second fit: NH2_CO vs WCO
coeffs2 = np.polyfit(WCO, NH2_CO_arr, 1)  # linear fit
fit_line2 = np.poly1d(coeffs2)
print(f"NH2_CO vs WCO: slope = {coeffs2[0]:.3e}, intercept = {coeffs2[1]:.3e}")

# Create a figure with 2 subplots (side by side)
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# First subplot: x-axis = NOH_arr
axs[0].scatter(NOH_arr, NH2_OH_arr, color='navy', label='NH2_OH', s=20)
axs[0].plot(NOH_arr, fit_line1(NOH_arr), color='skyblue', linestyle='--', label='Best fit', linewidth=1)
axs[0].set_xlabel('NOH [cm^-2]')
axs[0].set_ylabel('NH2 [cm^-2]')
axs[0].legend()
axs[0].grid(True)
axs[0].set_title('NH2 vs NOH')

# Second subplot: x-axis = WCO
axs[1].scatter(WCO, NH2_CO_arr, color='crimson', label='NH2_CO', s=20)
axs[1].plot(WCO, fit_line2(WCO), color='salmon', linestyle='--', label='Best fit', linewidth=1)
axs[1].set_xlabel('WCO [K km/s]')
axs[1].set_ylabel('NH2 [cm^-2]')
axs[1].legend()
axs[1].grid(True)
axs[1].set_title('NH2 vs WCO')

plt.tight_layout()
plt.show()
```

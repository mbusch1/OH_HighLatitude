```python
data = pd.read_csv('C:/Users/seven/Downloads/codarkfulldf3.csv')
```

```python
data = data.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
```

```python
# indexes for each line:
idx67 = np.where((data['line'] == '67'))[0]
idx65 = np.where((data['line'] == '65'))[0]
idxHI = np.where((data['line'] == 'HI'))[0]
idxCO = np.where((data['line'] == 'CO'))[0]
```

```python
fig, ax = plt.subplots()

ax.errorbar(x=data['FWHM (km/s)'][idx67], y=data['FWHM (km/s)'][idxCO], marker='o', ls='', label='CO', markersize = 4)
ax.errorbar(x=data['FWHM (km/s)'][idx67], y=data['FWHM (km/s)'][idxHI], marker='o', ls='', label='HI', markersize = 4)
ax.errorbar(x=data['FWHM (km/s)'][idx67], y=data['FWHM (km/s)'][idx65], marker='o', ls='', label='OH 1665', markersize = 4)
#ax.set_ylim([0.2, 4])
ax.set_xlim([0.2, 4])
ax.legend()
ax.set_ylabel("FWHM (km/s)")
ax.set_xlabel("OH 1667 FWHM (km/s)")
ax.set_title("FWHM Line Comparison - Total")
fig.tight_layout()
```

```python
fig, ax = plt.subplots()

ax.errorbar(x=data['T_peak (K)'][idx67], y=data['T_peak (K)'][idxCO], marker='o', ls='', label='CO', markersize = 3)
#ax.errorbar(x=data['T_peak (K)'][idx67], y=data['T_peak (K)'][idx65], marker='o', ls='', label='OH 1665', markersize = 4, alpha = 0.5)
plt.legend()
ax.set_xlabel("OH 1667 T peak (K)")
ax.set_ylabel("CO T peak (K)")
#ax.set_title("OH vs. CO T Peak")
```
